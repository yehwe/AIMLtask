# **FUNCTIONS**

* That's where Functions come in

## **Why do we need functions?**


Functions offer several benefits, including:

* **Modularity:** Functions allow you to break down complex tasks into smaller, more manageable pieces.
* **Reusability:** Once you've written a function to perform a particular task, you can call it as many times as you need without rewriting the same code.

* **Abstraction:** Functions hide the complexity of the tasks they perform.

* **Maintainability:** Code changes are easier when functionality is encapsulated in functions.

* **Testing and Debugging:** Functions can be tested individually, which makes it easier to identify and fix issues.

* **Parameterization:** Functions can take inputs, known as parameters or arguments.

**Note:** The name of the argument will not affect the name of the parameter

## Lambda Functions
* A lambda function is a small anonymous function.
*  "Anonymous" means it doesn't have a name like regular functions defined with def.
* Lambda functions are typically used for simple operations that can be expressed in a single line.
* A lambda function can take any number of arguments, but can only have one expression.

image.png

````python
#Multiplication of two arguments which are a and b, return the result:

x = lambda a, b : a * b


#call lambda
print(x(10, 5))
````

**comment**:

* `lambda:` This keyword indicates that we're defining a lambda function.
* a, b: These are parameters of the lambda function, just like function parameters. The lambda function expects two inputs when called.
*

## Why Use Lambda Functions?

The power of lambda is better shown when you use them as an anonymous function inside another function.

````python
def myfunc(n):
  return lambda a : a * n
````

````python
v = myfunc( n = 5)
````

````python
type(v)
````

````python
#example map
numbers = [1,2,3,4,5,6]
squared_numbers = map(lambda x: x**2, numbers)
print(list(squared_numbers))

#example of filter
numbers = [1,2,3,4,5,6]

even_numbers = filter(lambda x: x%2 ==0, numbers)

print(list(even_numbers))
````

##  **While Loop**


The while loop executes a block of code as long as its **condition** is **True**.
It's essential to modify the loop control variable inside the loop to ensure that the loop condition eventually becomes **False**, or else you'll have an **infinite loop**.

* while loop is like an if-statement that repeats

* Loops have iteration variables that change each time through the loop



**The Basic Syntax**
```
while condition:
    # code to execute while the condition is True
```
* In the while loop, expression/condition is checked first.
* The body of the loop is entered only if the expression/condition evaluates to True.
* After one iteration, the expression/condition is checked again. This process continues until the test_expression evaluates to False.

**Key Point**:
The while loop is used when:

* The number of iterations is determined by a condition rather than a predetermined count.
* You want to loop indefinitely until a certain condition is met.


**Example 1**

````python
# Example 1: Print numbers less than 8
count = 1

while count < 8:

  print(count)

  count = count + 1

#code continues
print('The while is over')

````

**Example 2:**
Here is a simple example, showing how a game will stay active as long as the player has enough power.

````python
# The player's power starts out at 5.
power = 5

# The player is allowed to keep playing as long as their power is over 0.
while power > 0:
    print("You are still playing, because your power is %d." % power)
    # Your game code would go here, which includes challenges that make it
    #   possible to lose power.
    # We can represent that by just taking away from the power.
    power = power - 1

print("\nOh no, your power dropped to 0! Game Over.")
````

## "Infinite" Loops and Break

An infinite loop runs indefinitely because its condition always remains True.
They are generally considered errors but can be used intentionally in some cases.
The **break** statement ends the current loop, and jumps to the statement immediately following the looo

````python
n = 1
while n > 0:  # This creates an infinite loop
    print('Python')
    print('C++')
    n += 1

    if n > 3:
        break  # This will break out of the loop when count is greater than 2

print('Hey stop the loop')
````

## **Using continue**

The continue statement skips the rest of the loop's current iteration and moves to the next iteration.
Useful for skipping certain iterations based on a condition.

**Example:**

````python
#Print numbers from 1 to 10, except 4 and 7.
num = 1

while num <= 10:

  if num == 4 or num == 7:
    num += 1
    continue

  print(num)
  num+=1





````

## **While if-else Loop**

A while loop can have an optional if-else block.
* We use if-else statement in the loop when conditional iteration is needed. i.e., If the condition is True, then the statements inside the if block will execute othwerwise, the else block will execute.

* If the loop completes without encountering a break statement, the code in the else block is executed.
* If a break statement is encountered inside the while loop, the else block is skipped.



## **Example: Number Guessing Game**

Let's create a simple game where the user has to guess a pre-defined number.



image.png

[Image source](https://www.google.com/url?sa=i&url=https%3A%2F%2Fwww.teachwithict.com%2Fnumberguess.html&psig=AOvVaw1WapP-2-S89Kko4jH9hqoj&ust=1692554238723000&source=images&cd=vfe&opi=89978449&ved=0CBIQjhxqFwoTCPjykPul6YADFQAAAAAdAAAAABAE)

# **For Loop (Definite Loop)**

The for loop is used to iterate over a sequence (like a list, tuple, dictionary, or string) or other iterable objects.
Has a definite number of iterations.

**Key Point:** The for loop is used when:

* You know beforehand how many times you want to iterate.
* You want to iterate over the items of a sequence (like a list, tuple, string, dictionary, or set).

**Example of iterating a specific number of times:**

`range()` generate a list to loop over "on the fly".

````python
#Print numbers from 1 to 3:
#n-1
for i in range(1, 10 , 2):
    print(i)
````

**Example of iterating over a sequence (list in this case):**

````python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
````

**Enumerate**
You can achieve the same effect as above by using the enumerate() function, to get two values each step through the loop.

````python
# the enumerate function will return two values each time the loop statement is run:
# "indx" will be the current index value, and "val" will be the current element of my_list
for indx, val in enumerate(fruits):
    print(indx, val)
````

**Loop through the characters in a string**

````python
s = "One at a time"
for character in s:
    print(character)
````

## Loop Patterns (counting, summing, averaging, etc.)

Loops can be used to perform various operations on data.
Common patterns include counting items, summing values, calculating averages, finding maximum/minimum values, etc.

````python
#Counting - Count how many times the number '2' appears in a list:
numbers = [1, 2, 3, 2, 4, 2, 5]
count = 0
for num in numbers:
    if num == 2:
        count += 1
print(count)
````

````python
#Summing - Calculate the sum of all numbers in a list:

numbers = [1, 2, 3, 4, 5]
total = 0
for num in numbers:
    total += num
    print(total)
print(total)
````

Averaging - Calculate the average of numbers in a list:

````python
numbers = [1, 2, 3, 4, 5]
total = 0
count = 0
for num in numbers:
   total += num
   count += 1

print('Exit loop')
average = total / count
print(average)

#alternative code
#total = sum(numbers)
#average = total / len(numbers)
````

````python
total = sum(numbers)
average = total / len(numbers)

print(average)
````
